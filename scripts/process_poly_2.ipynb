```python
with open('./VID_kirinuki_torim.mp4.json') as f:
    dat = json.load(f)
```

```python
dat = dat['frames']
```

```python
dat
```

```python
mask_dat= []

for key,ary in dat.items():
#     print(key,val)
    img = np.zeros((height,width,3), np.uint8)
    img.fill(255)

    for val in ary:
        uid = val['UID']
        x1 = int(val['x1'])
        x2 = int(val['x2'])
        y1 = int(val['y1'])
        y2 = int(val['y2'])
#         print(x1,x2,y1,y2)
        img = cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,0),1)
    
    img = img.transpose((1,0,2))
    img = img.tolist()
    img_p = []
    for row in img:
        row_p = []
        bi_px = 0
        for px in row:
            if px==[0,0,0]:
                bi_px = 1
            row_p.append(bi_px)
        img_p.append(row_p)
    
    img_p = np.array(img_p)
    img_p = img_p.transpose((1,0))
    

    mask_dat.append({
        'UID':uid,
        'img':img_p.tolist()
    })
```
